## Cables

- USB-2.0 micro-B
  - El estándar fue pensado para 5v@100mA de alimentación inicial y 5v@500mA luego del "negociado". La mayoría de cargadores ignoran estos límites, pero los cables son los mismos
  - Los conectores USB-2.0 micro-B son fabricados, en general, para soportar 1.8A max en los pines de alimentación, con una resistencia de contacto de 30mΩ max

&nbsp;

- USB-C
  - La corriente mínima tras negociado es de 900mA.
  - Los cables de USB-A 2.0 a USB-C por norma deben admitir 3A de corriente.
  - La máxima caída de tensión para la línea VBus, incluyendo contactos es de 500mV; para GND es 250mV.
  - Los conectores USB-C admiten entre 3A y 5A, con una resistencia de contacto de 30mΩ (según algunos fabricantes)

## Medicion de ripple y prueba de stress:

### Raspberry Pi 3b+

- Resistencias de potencia de 1.7Ω  y 2.7Ω para la prueba de ripple.
- `stress-ng -c 4 -l 95 -t 30` para la prueba de stress. Se indica min/max de un multímetro en modo captura.
- Display 7" HDMI alimentado por puerto USB como demanda adicional de corriente.
- La columna "USB" indica que se alimenta la Raspberry por USB, ya sea por el cable propio de la fuente, o el color azul mencionado, de 268mΩ (resistencia del conductor rojo)
  - ⛔ Indica que el cable USB ya está incorporado en la fuente y esta prueba seria inválida
- La columna "+5V" indica alimentación por medio del correspondiente pin mediante un cable armado
- ➡️ indica que no es necesaria esta prueba, al pasar la prueba por USB

Un efecto secundario de este capacitor, es propiciar una fuga de AC de 50hz, cuyo valor de tensión dependerá de los valores del divisor de tensión capacitivo formado entre este capacitor y el propio del formado entre los bobinados primario y secundario.

Esta tensión estará presente siempre que se use una fuente switching, con la Raspberry Pi flotante, es decir, con GND sin conexión a tierra; y bajo ciertas condiciones, puede ocasionar fallas, como resetearse al tocar el pin RUN.
Este problema se elimina aterrizando la R-Pi, con una conexión directa a tierra o a través de un capacitor de 1uF. Esta conexión también solucionará problemas de ESD, que también pueden resetear la R-Pi o peor.

## Otros componentes

### Capacitores electrolíticos

La mayoría de los encontrados en estas fuentes, son de electrolito líquido. Los foil de aluminio (ánodo y cátodo) son separados por un papel impregnado en electrolito líquido. Al calentarse por encima de la temperatura máxima para la cual están fabricados, el electrolito comienza a ganar presión de vapor, y el armazón del capacitor se expande, liberándose parte del electrolito tanto por el sello de goma inferior, como por las ranuras de seguridad marcadas en la parte superior. Esta pérdida de electrolito aumenta el ESR, lo que aumenta la temperatura bajo corrientes de ripple, lo que continúa con este efecto bola de nieve, hasta que el capacitor pierde por completo su función.
